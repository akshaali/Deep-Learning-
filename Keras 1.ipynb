## Procesing Data 

```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler 
```

```python
train_lables = []
train_sample = []
```

Example Data 
* An experimental drug was tested from individual ages between 13 to 65.
* The trial had 2100 participants. Half were under 65 and half were above 65 .
* 95% of patients 65 or older have experinced side effects .
* 95% of patients below than 65 haven't experienced side effects .

### Getting raw data :


```python
for i in range(1000):
    random_younger = randint(13,64)
    train_sample.append(random_younger)
    train_lables.append(0)
    
    random_older = randint(65,100)
    train_sample.append(random_older)
    train_lables.append(1)

for i in range(50):
    random_younger = randint(13,64)
    train_sample.append(random_younger)
    train_lables.append(1)
    
    random_older = randint(65,100)
    train_sample.append(random_older)
    train_lables.append(0)
```

```python
#print raw data 
for i in train_sample:
    print(i)
```

```python
for i in train_lables:
    print(i)
```

```python
train_sample = np.array(train_sample)
train_lables = np.array(train_lables)
```

```python
scaler = MinMaxScaler()
scaled_train_sample = scaler.fit_transform((train_sample).reshape(-1,1))
```

```python
scaled_train_sample
```

```python
1+9
```

```python
import keras
from keras import backend as k
from keras.models import Sequential
from keras.layers  import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
#model = Sequential([l1,l2,l3])
model = Sequential([Dense(16, input_shape=(1,), activation='relu'),
                  Dense(32, activation='relu'), Dense(2, activation='softmax')])
```

```python
model.summary()
```

```python
model.compile(Adam(lr=.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

### Validation set 

```python
model.fit(scaled_train_sample, train_lables,validation_split=0.1, batch_size=10, epochs=20, shuffle=True, verbose=2)
```

### Prediction Set

```python
test_sample = []
test_lables = []
```

```python
for i in range(200):
    random_younger = randint(13,64)
    test_sample.append(random_younger)
    test_lables.append(0)
    
    random_older = randint(65,100)
    test_sample.append(random_older)
    test_lables.append(1)

for i in range(10):
    random_younger = randint(13,64)
    test_sample.append(random_younger)
    test_lables.append(1)
    
    random_older = randint(65,100)
    test_sample.append(random_older)
    test_lables.append(0)
```

```python
test_sample = np.array(test_sample)
```

```python
scaled_test_sample = scaler.fit_transform((test_sample).reshape(-1,1))
```

```python
prediction = model.predict(scaled_test_sample, batch_size= 10, verbose=0)
```

```python
for i in prediction:
    print(i)
```

```python
rounded_prediction = model.predict_classes(scaled_test_sample, batch_size=10, verbose=0)
```

```python
for i in rounded_prediction:
    print(i)
```
